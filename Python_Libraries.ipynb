# Activity **1.2**

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Score': [85, 90, 78, 92]}
df = pd.DataFrame(data)
print(df)

print("Data Frame Coloums", df.columns)
print("Data Frame Shape (rows, cols)", df.shape)
print("Data Frame Coloums (types): ", df.dtypes)
```

```python
print(df.loc[:,"Age"])
```

```python
df[df["Score"] > 80]
```

# Activity **2.1**

```python
df[df["Age"] > 30].sort_values(by="Age")
```

# Activity **2.2**

```python
df["Grade"] = df["Score"].apply(lambda score:
    "A" if score >= 90 else
    "B" if score >= 80 else
    "C" if score >= 70 else
    "D" if score >= 60 else "F"
)
print(df)
```

```python
df.drop(columns=["Grade"], inplace=True)
print(df)
```

# **Lab**-**2** *Matplotlib*





# Activity **1.1**

```python
import matplotlib.pyplot as plt
# Sample data
x = np.linspace(0, 10, 100)
y = np.cos(x)
# Create plot
plt.plot(x, y, label="Sine Wave")
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

# Activity **1.2**

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
x = np.linspace(0, 10, 100)
ax[0].plot(x, np.sin(x), 'r', label="Sine Wave")
ax[1].plot(x, np.cos(x), 'b', label="Cosine Wave")
ax[0].set_title("Sine Function")
ax[1].set_title("Cosine Function")
for a in ax:
    a.set_xlabel("X-axis")
    a.set_ylabel("Y-axis")
    a.legend()
    a.grid(True)
plt.tight_layout()
plt.show()
```

# Activity **1.3**

```python
plt.plot(x, np.sin(x), linestyle='--', marker='s', color='r', label='Sine')
plt.plot(x, np.cos(x), linestyle='-.', marker='o', color='y', label='Cosine')
plt.legend()
plt.title("Customized Line Styles")
plt.show()
```

# Activity **3.1**

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasets/s-and-p-500/master/data/data.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

print(df.columns)

# Calculate a 30-day rolling average of the SP500 values
df['Rolling Average'] = df['SP500'].rolling(window=30).mean()

plt.figure(figsize=(10, 5))
plt.plot(df.index, df['SP500'], color='b', linestyle='-', label="S&P 500")
plt.plot(df.index, df['Rolling Average'], color='orange', linestyle='--', label="30-Day Rolling Average")
plt.xlabel("Date")
plt.ylabel("S&P 500 Value")
plt.title("S&P 500 Index with Rolling Average")
plt.legend()
plt.show()
```

# Activity **3.2**

```python
labels = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Swift', 'Go']
sizes = [30, 20, 15, 10, 8, 7, 6, 4]
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta', 'yellow']
explode = (0.1, 0, 0, 0, 0, 0, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, shadow=True, startangle=140)
plt.title("Programming Language Popularity")
plt.show()
```

# Activity **4**

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]
data = np.random.randn(500)

fig, ax = plt.subplots(2, 2, figsize=(10, 6))

ax[0, 0].plot(x, y, 'r')
ax[0, 0].set_title("Sine Wave")
ax[0, 1].bar(categories, values, color='b')
ax[0, 1].set_title("Bar Chart")
ax[1, 0].scatter(x, y, c='g')
ax[1, 0].set_title("Scatter Plot")
ax[1, 1].hist(data, bins=20, color='purple')
ax[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```

# Activity **1.1**

```python
import seaborn as sns

tips = sns.load_dataset("tips")
print(tips.head())
print(tips.info())
print(tips.describe())
sns.set_style("whitegrid")
```

# Activity **2**

```python
sns.histplot(tips['total_bill'], kde=True)
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.barplot(x='day', y='tip', data=tips)
plt.show()
```

# Activity **3**

```python
corr = tips.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
g = sns.FacetGrid(tips, col="sex")
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip')
plt.show()
```

# Activity **4**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tips, palette='pastel')
plt.show()
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.legend(title="Gender")
plt.show()
```
